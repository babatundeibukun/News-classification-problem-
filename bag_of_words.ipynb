```python
corpus = [
    'This is the first document.',
     'This document is the second document.',
     'And this is the third one.',
     'Is this the first document?',
 ]
vectorizer = CountVectorizer(ngram_range=(2, 2))
X = vectorizer.fit(corpus)
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")

def preprocess(text):
  doc=nlp(text)

  filtered_words = []

  for words in doc:
    if words.is_stop or words.is_punct:
      continue
    filtered_words.append(words.lemma_)
  return " ".join(filtered_words)


```

```python
preprocess('This is a big boy who stays committed to his goals.')
```

```python
preprocessed_corpus = [preprocess(text) for text in corpus]
preprocessed_corpus
```

```python
df_balanced=pd.concat([df_politics,df_wellness,df_entertainment,df_travel,df_style,df_parenting,df_health,df_black,df_religion], axis=0)
df_balanced
```

```python
df_balanced.category.value_counts()
```

```python
target = {'POLITICS':0, 'WELLNESS' :1, 'ENTERTAINMENT':3,'TRAVEL':4, 'STYLE':5, 'PARENTING':6, 'HEALTHY LIVING': 7, 'BLACK VOICES': 8, 'RELIGION': 9 }
df_balanced['category_num']= df_balanced.category.map(target)
```

```python
df_balanced.sample(5)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df_balanced.headline,
    df_balanced.category_num,
    test_size=0.2,
    random_state=222,
    stratify=df_balanced.category_num
)
```

```python
X_train.head()
```

```python
y_train.value_counts()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report


clf=Pipeline([
    ('bow_vectorizer', CountVectorizer(ngram_range=(1,2))),
    ('Model', MultinomialNB())
])

#fitting the model
clf.fit(X_train, y_train)

#prediction
y_preds=clf.predict(X_test)

print(classification_report(y_test, y_preds))
```
